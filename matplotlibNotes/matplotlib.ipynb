## matplotlib折线图


```python
#绘制一天的气温变化图
import matplotlib.pyplot as plt

y = [15, 13, 14, 5, 17, 20, 25, 26, 26, 24, 22, 18];
x = range(2, 26, 2)
#设置图片大小
plt.figure(figsize=(20,8), dpi=80)
plt.plot(x, y); #利用传入的x和y，通过plot绘制出折线图
#绘制x，y的刻度
x_ticklabels = [i/2 for i in range(4, 49)]
plt.xticks(x_ticklabels);
#设置x，y标签
plt.xlabel("time");
plt.ylabel("temperature °C");
#设置标题
plt.title('temperature varing with time t')
plt.show();
#图片保存
# plt.savefig('./temp.svg')
```

```python
#表示10点到12点的每一分钟的气温
import random
# import matplotlib
x = range(0, 120);
a = [random.randint(20, 35) for i in range(0, 120)];

plt.figure(figsize = (20, 8))
plt.plot(x,a);
#调整x轴的刻度
_x = list(x);
_xtick_labels = ["10点{}分".format(i) for i in range(60)];
_xtick_labels += ["11点{}分".format(i) for i in range(60)];
#取步长，数字和字符串一一对应，数据的长度一样
#同时设置中文显示
#设置x，y标签
plt.xlabel("时间");
plt.ylabel("温度 °C");
#设置标题
plt.title('不同时间下的温度')
plt.xticks(_x[::3], _xtick_labels[::3],  rotation = 45); #rotation旋转的度数
```

```python
#绘制一个新的折线图
number1 = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 5, 4, 3, 3, 1, 1, 1];
number2 = [0 , 1 , 3, 5, 6, 7, 3, 2, 5, 6, 7, 1, 2, 5, 7, 4, 2, 1, 1, 1];
# print(len(number1))

ages = range(11, 31);

plt.figure(figsize = (20, 8));
plt.plot(ages, number1, label = '小明');
plt.plot(ages, number2, label = '小李');
_x_ticks = ["在{}岁".format(i) for i in range(11, 31)];
plt.legend();#显示图例
plt.xticks(ages, _x_ticks, rotation = 45);
plt.xlabel('年龄');
plt.ylabel('对象个数');
plt.grid()#显示网格
plt.title('不同年龄段，交的朋友个数');
plt.show();
```
