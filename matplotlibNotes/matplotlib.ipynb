```python
#
import matplotlib.pyplot as plt

y = [15, 13, 14, 5, 17, 20, 25, 26, 26, 24, 22, 18];
x = range(2, 26, 2)
#
plt.figure(figsize=(20,8), dpi=80)
plt.plot(x, y); #xyplot
#xy
x_ticklabels = [i/2 for i in range(4, 49)]
plt.xticks(x_ticklabels);
#xy
plt.xlabel("time");
plt.ylabel("temperature C");
#
plt.title('temperature varing with time t')
plt.show();
#
# plt.savefig('./temp.svg')
```

```python
#1012
import random
# import matplotlib
x = range(0, 120);
a = [random.randint(20, 35) for i in range(0, 120)];

plt.figure(figsize = (20, 8))
plt.plot(x,a);
#x
_x = list(x);
_xtick_labels = ["10{}".format(i) for i in range(60)];
_xtick_labels += ["11{}".format(i) for i in range(60)];
#
#
#xy
plt.xlabel("");
plt.ylabel(" C");
#
plt.title('')
plt.xticks(_x[::3], _xtick_labels[::3],  rotation = 45); #rotation
```
